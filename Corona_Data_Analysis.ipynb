### Here I download data from John Hopkins Covid Data Github page

## Locate rows
`loc[]` locates the rows (slices it, but i forgot what this means specifically
- `loc` has a lot of ways to call it

```python
daily_df.loc[daily_df.Country_Region == 'Germany']
# daily_df.loc[226]
```

## Creating a new data frame from sliced one 
And rename last rows State name 

```python
germany = pd.DataFrame(daily_df.loc[daily_df.Country_Region == 'Germany'])
germany.rename(columns={'Province_State':'Bundesland'}, inplace=True)
germany['Bundesland'] = germany['Bundesland'].replace(['Unknown'],'Kreuzfahrtschiff')
germany
```

### Matplot style plotting

```python
plt.figure(figsize=(12,8))
plt.bar(germany.Bundesland, germany.Incident_Rate )
plt.xticks(rotation=90)
plt.ylabel("Cases - cumulative total per 100000 population")
plt.show()
```
